```python
from src import k_means
from sklearn.cluster import KMeans
from pandas import read_csv
import plotly_express as px
import numpy as np
```

```python
cols = ["kdRatio","timePlayed"]
cod_df = read_csv("data/cod.csv", usecols=cols)
cod_df = cod_df[(cod_df["kdRatio"]>0) & (cod_df["timePlayed"]>100) & (cod_df["kdRatio"]<1.5) & (cod_df["timePlayed"] < 2500)]
```

```python
X = np.array(cod_df)
sklearn_kMeans_clstr = KMeans(n_clusters=3, random_state=1)
sklearn_kMeans_clstr.fit(cod_df)
cod_df['labels'] = sklearn_kMeans_clstr.labels_.astype('int32')
fig = px.scatter(cod_df, x='timePlayed', y='kdRatio', color='labels')
fig.show() 
```
